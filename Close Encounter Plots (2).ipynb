```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
import seaborn as sns
os.chdir("C:\Research")
```

```python
#######################################################
### TIDAL EVOLUTION WITHOUT COLLISIONLESS ENCOUNTERS ##
#######################################################

tide_aei = pd.read_fwf('tide_aei.txt',header=None,names=['Time','Semimajor axis','Eccentricity','Inclination'])
```

```python
# Semimajor axis (just tidal evolution)

tide_res = []
for value in tide_aei['Semimajor axis']:
    re = value * 23481.065876628
    tide_res.append(re)
    
fig, ax = plt.subplots()
plt.tick_params(axis='both',labelsize=14)
   
plt.scatter(x=tide_aei['Time'],y=tide_res,s=20,c='slategray')
plt.ylim(0.00,20)
plt.xlim(0,200000)
plt.xlabel("Time (yrs)",fontsize=20,fontname='Arial')
plt.ylabel("Semimajor axis (earth radii)",fontsize=20,fontname='Arial')
plt.title("Tidal evolution of semimajor axis over time",fontsize=24,fontname='Arial')
plt.rcParams['figure.figsize']=[20,10]
fig.savefig('C:\\Research\\DPS_figures\\tide_a.png',quality=95)
plt.show()
```

```python
# Eccentricity (just tidal evolution) with Ks = 0.3

fig, ax = plt.subplots()
plt.tick_params(axis='both',labelsize=14)

plt.scatter(x=tide_aei['Time'],y=tide_aei['Eccentricity'],s=10,c='indianred')
plt.ylim(0.054,0.056)
plt.xlim(0,200000)
plt.xlabel("Time (yrs)",fontsize=20,fontname='Arial')
plt.ylabel("Eccentricity",fontsize=20,fontname='Arial')
plt.title("Tidal evolution of eccentricity over time",fontsize=24,fontname='Arial')
plt.rcParams['figure.figsize']=[20,10]
fig.savefig('C:\\Research\\DPS_figures\\tide_e.png',quality=95)
plt.show()
```

```python
inc_deg_tides =[]
for inc in tide_aei['Inclination']:
    inc_deg_tides.append(inc*57.2958)
```

```python
# Inclination (just tidal evolution)
fig, ax = plt.subplots()
plt.tick_params(axis='both',labelsize=14)

plt.scatter(x=tide_aei['Time'],y=inc_deg_tides,s=10,c='orange')
plt.ylim(5.141,5.146)
plt.xlim(0,200000)
plt.xlabel("Time (yrs)",fontsize=20,fontname='Arial')
plt.ylabel("Inclination (degrees)",fontsize=14,fontname='Arial')
plt.title("Tidal evolution of inclination over time",fontsize=24,fontname='Arial')
plt.rcParams['figure.figsize']=[20,10]
fig.savefig('C:\\Research\\DPS_figures\\tide_i.png',quality=95)
plt.show()
```

```python
xearth = []
for value in Earth['x']:
    xearth.append(value)
xearth = np.asarray(xearth)

yearth = []
for value in Earth['y']:
    yearth.append(value)
yearth = np.asarray(yearth)

xmoon = []
for value in Moon['x']:
    xmoon.append(value)
xmoon = np.asarray(xmoon)

ymoon = []
for value in Moon['y']:
    ymoon.append(value)
ymoon = np.asarray(ymoon)

x = xearth-xmoon
y = yearth-ymoon

print(x)
print(y)

plt.scatter(x,y,s=35)
plt.xlim(-0.0002,0.0)
plt.ylim(-0.00035,-0.0002)
plt.rcParams['figure.figsize']=[15,15]
plt.show()
```
